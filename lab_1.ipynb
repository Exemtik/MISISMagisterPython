### Вычислить	сумму	s,	прекращая	суммирование,	когда	очередной	член	суммы	по	абсолютной	величине	станет	меньше	 0,0001,	при	изменении	аргумента	x в	указанном	диапазоне	[а, b]	c	шагом	h. Для	сравнения	в	каждой	точке	вычислить	также	функцию	y = f(x),	являющуюся	аналитическим	выражением	ряда.

```python
import numpy as np
EPS = 0.0001

def task_2(a, b, h):
    for x in np.arange(a, b+h, h):

        i = 1
        s = 0
        p = x
        while abs(x)**i >= EPS:         
            s += p * np.sin(i*np.pi/4)
            i += 1
            p *= x
            
        y = (x*np.sin(np.pi/4)) / (1 - 2*x*np.cos(np.pi/4) + x**2)
        print(f"x={x:.1f}  s~~{s:.8f}  y={y:.8f} diff={abs(s-y):.2f}  i={i}")

a = 0.1
b = 0.8
h = 0.1
task_2(a, b, h)
```

```python
import numpy as np
EPS = 0.0001

def task_5(a, b, h):
    for x in np.arange(a, b+h, h):

        i = 1
        s = 0
        while True:
            tmp_x = (-1)**i * np.cos(i*x)/i**2          
            s += tmp_x
            if abs(tmp_x) < EPS:
                break
            i += 1

        y = (x**2 - (np.pi**2)/3) / 4
        print(f"x={x:.3f}  s~~{s:.8f}  y={y:.8f}, diff={abs(s-y):.2f}  i={i}")

a = np.pi/5
b = np.pi
h = np.pi/25
task_5(a, b, h)
```

```python
import numpy as np
EPS = 0.0001

def factorial(val):
    res = 1
    for n in range(2, val+1):
        res *= n
    return res
    

def task_7(a, b, h):
    for x in np.arange(a, b+h, h):

        i = 0
        s = 0
        while True:
            tmp_x = x**(2*i) / factorial(2*i)
            s += tmp_x
            if abs(tmp_x) < EPS:
                break
            i += 1

        y = (np.e**x + np.e**(-x)) / 2
        print(f"x={x:.2f}  s~~{s:.8f}  y={y:.8f} diff ={abs(s-y):.2f} i={i}")

a = 0.1
b = 1
h = 0.05
task_7(a, b, h)
```

### Решить	задачи	II	уровня	для	случая,	когда	количество	данных	заранее	неизвестно.	В	программе	необходимо	обеспечить	прекращение	ввода,	как	только	входной	поток	иссякнет.

* 10

```python
import math

def task10(students):
    count_good = 0
    for grades in students:
        if 2 not in grades and 3 not in grades:
            count_good += 1
    return count_good


def task11(students):
    bad_students = 0
    total_sum = 0
    total_count = 0
    for grades in students:
        total_sum += sum(grades)
        total_count += len(grades)
        if 2 in grades:
            bad_students += 1
    avg = total_sum / total_count if total_count > 0 else 0
    return bad_students, round(avg, 2)


def task12(shape, values):
    areas = []
    for r in values:
        if shape == "square":
            areas.append(r*r)
        elif shape == "circle":
            areas.append(math.pi*r*r)
        elif shape == "triangle":
            areas.append((math.sqrt(3)/4)*r*r)
    return areas


import math

def test_task10():
    students = [[5,5,4,4], [4,3,5,4]]
    if(task10(students) == 1):
        print("Task passed Successfully")

def test_task11():
    students = [[5,5,5,5], [4,2,5,3]]
    bad, avg = task11(students)
    if(bad == 1):
        print("Task passed Successfully")
    if(math.isclose(avg, 4.25, rel_tol=1e-9)):
        print("Task passed Successfully")

def test_task12():
    if(task12("square", [2,3]) == [4,9]):
        print("Task passed Successfully")
    if([round(x,4) for x in task12("circle", [1])] == [round(math.pi,4)]):
        print("Task passed Successfully")
    if([round(x,4) for x in task12("triangle", [2])] == [round(math.sqrt(3),4)]):
        print("Task passed Successfully")

test_task10()
test_task11()
test_task12()
```
