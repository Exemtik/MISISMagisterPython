# Уровень 2


```python
import pandas as pd

# Таблица сотрудников
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4, 5, 6],
    'name': ['Анна', 'Борис', 'Виктор', 'Галина', 'Дмитрий', 'Елена'],
    'dept_id': [101, 102, 101, 103, 102, 101],
    'salary': [50000, 60000, 55000, 70000, 65000, 45000]
})

# Таблица отделов
departments = pd.DataFrame({
    'dept_id': [101, 102, 103, 104],
    'dept_name': ['IT', 'HR', 'Finance', 'Marketing']
})

# Таблица городов
cities = pd.DataFrame({
    'emp_id': [1, 2, 3, 4, 5, 6],
    'city': ['Москва', 'Санкт-Петербург', 'Москва', 'Новосибирск', 'Санкт-Петербург', 'Москва']
})

print("Сотрудники:")
display(employees)
print("\nОтделы:")
display(departments)
print("\nГорода:")
display(cities)
```

```python
merged_df = pd.merge(pd.merge(employees, departments, on="dept_id", how='inner'), cities, on="emp_id", how='inner')
merged_df
```

```python
merged_df.loc[merged_df['dept_name'] == 'IT']
```

# Уровень 3


```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Создаем расширенный датасет
np.random.seed(42)

employees = pd.DataFrame({
    'employee_id': range(1, 21),
    'first_name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie', 'Diana', 'Eve', 'Frank', 
                   'Grace', 'Henry', 'Ivan', 'Julia', 'Kevin', 'Laura', 'Mike', 'Nancy',
                   'Oliver', 'Paula', 'Quinn', 'Rachel'],
    'last_name': ['Smith', 'Doe', 'Johnson', 'Brown', 'Wilson', 'Lee', 'Davis', 'Miller',
                  'Taylor', 'Anderson', 'Clark', 'Harris', 'Martin', 'Thompson', 'Garcia',
                  'Martinez', 'Robinson', 'Walker', 'Young', 'Allen'],
    'department': ['IT']*5 + ['HR']*4 + ['Finance']*4 + ['Marketing']*3 + ['Sales']*4,
    'salary': np.random.randint(45000, 120000, 20),
    'hire_date': pd.date_range('2018-01-01', periods=20, freq='ME'),
    'city': ['New York']*6 + ['Chicago']*5 + ['Boston']*4 + ['Seattle']*5,
    'age': np.random.randint(22, 55, 20),
    'manager_id': [None, 1, 1, 2, 3, None, 6, 6, 7, 8, None, 11, 11, 12, 13, None, 16, 16, 17, 18],
    'bonus': np.random.randint(0, 10000, 20)
})

# Добавляем пропуски для реалистичности
employees.loc[[2, 8, 15], 'bonus'] = np.nan
employees.loc[[5, 12], 'city'] = None

print("Расширенный датасет сотрудников:")
display(employees.head(8))
```

```python
mean_salaries = employees.groupby('department').agg({'salary' : "mean"})
mean_salaries
```

```python
employees
```

```python
for index, row in employees.iterrows():
    employees.loc[index, 'dif_salaries'] = (row['salary'] - mean_salaries.loc[row['department']]).values

employees
```

```python
higher_average = employees.loc[employees['dif_salaries'] > 0]
```

```python
higher_average[['first_name', 'department', 'salary', 'dif_salaries']]
```
