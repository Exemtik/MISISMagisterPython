```python
def lab2_task1(lst):
    if not lst:
        return None
    index_lst = []
    max_elem = lst[0]
    for i, elem in enumerate(lst):
        if elem > max_elem:
            max_elem = elem
            index_lst.clear()
            index_lst.append(i)
        elif elem == max_elem:
            index_lst.append(i)
        else:
            continue
    return index_lst

lst = [1, 2, 3, 5, 5, 3, 2, 5]
max_index = lab2_task1(lst)
print(max_index)

def lab2_task1_matrix(matrix):
    if not matrix or not matrix[0]:
        return []

    max_elem = matrix[0][0]
    max_indices = [(0, 0)]

    for i, row in enumerate(matrix):
        for j, val in enumerate(row):
            if val > max_elem:
                max_elem = val
                max_indices = [(i, j)]
            elif val == max_elem:
                max_indices.append((i, j))

    return max_elem, max_indices

matrix = [
    [1, 5, 3],
    [5, 2, 5],
    [0, 4, 5]
]

max_value, positions = lab2_task1_matrix(matrix)
print(positions)
```

```python
def lab2_task2(lst):
    if not lst:
        return lst

    max_elem = lst[0]
    max_indices = [0]

    for i in range(1, len(lst)):
        val = lst[i]
        if val > max_elem:
            max_elem = val
            max_indices = [i]
        elif val == max_elem:
            max_indices.append(i)

    for k, idx in enumerate(max_indices, start=1):
        lst[idx] += k

    return lst


lst = [1, 2, 3, 5, 5, 3, 2, 5, 5, 5]
print(lab2_task2(lst), "\n")

def lab2_task2_matrix(matrix):
    if not matrix or not matrix[0]:
        return matrix

    max_elem = matrix[0][0]
    max_positions = [(0, 0)]

    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            val = matrix[i][j]
            if val > max_elem:
                max_elem = val
                max_positions = [(i, j)]
            elif val == max_elem:
                max_positions.append((i, j))

    for k, (i, j) in enumerate(max_positions, start=1):
        matrix[i][j] += k

    return matrix

matrix = [
    [1, 5, 3],
    [5, 2, 5],
    [0, 4, 5]
]
result = lab2_task2_matrix(matrix)
for row in result:
    print(row)
```

```python
def lab2_task3(lst):
    pos_index = 0  
    for i in range(len(lst)):
        if lst[i] >= 0:
            lst[pos_index], lst[i] = lst[i], lst[pos_index]
            pos_index += 1
    return lst

lst = [3, -1, 0, -7, 5, -2]
print(lab2_task3(lst), "\n")

def lab2_task3_matrix(matrix):
    for row in matrix:
        pos_index = 0
        for i in range(len(row)):
            if row[i] >= 0:
                row[pos_index], row[i] = row[i], row[pos_index]
                pos_index += 1
    return matrix

matrix = [
    [3, -1, 0],
    [-2, 5, -3]
]
for row in lab2_task3_matrix(matrix):
    print(row)
```

```python
def sum_range(a, b):
    if a > b:
        a, b = b, a 
    result = 0
    for i in range(a, b + 1):
        result += i
    return result

x = int(input())
y = int(input())
print(sum_range(x, y))
```

```python
def max_of_three(a, b, c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val

first_val = int(input())
second_val = int(input())
third_val = int(input())
print(max_of_three(first_val, second_val, third_val))
```

```python
def product(lst):
    result = 1
    for x in lst:
        result *= x
    return result

lst = list(map(float, input().split()))
print(product(lst))
```
